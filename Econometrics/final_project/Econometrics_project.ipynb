### 1) total expense -> target variable

```python
# 각 bmi 기준 별로 변수들이 총 지출에 얼마만큼 영향을 미치는지 추정
# demo_socio 변수에는: age, educ, srhealth 제거 -> 다중공선성 줄었음
# full model 변수에는: 전부
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

demo_socio_formula = 'ln_totalexp ~ ln_income  +  male'
additional_formula = ' + high_bp + high_chol + phy_lim + diabetes + chd '
```

```python
bmi_list = [bmi_0,bmi_1,bmi_2,bmi_3]
for bmi in bmi_list:
    print('---------------------------------------------------------------------------------------------')
    print('bmi')
    full_model = ols(demo_socio_formula+additional_formula, data=bmi) #bmi_바꿔주면서 적용
    full_result = full_model.fit()
    print('full_result',full_result.summary())
    print('---------------------------------------------------------------------------------------------')
```

### 2) Healthcare utilization -> target variable

## Poisson model

```python
from patsy import dmatrices
import statsmodels.discrete.discrete_model as dm
import statsmodels.api as sm


bmi_list = [bmi_0,bmi_1,bmi_2,bmi_3]
for bmi in bmi_list:
    print('---------------------------------------------------------------------------------------------')
    print('bmi')
    #dpendent variable: dr_visits

    # 포아송 모델에서는 totalexp, income 스케일링 x
    demo_socio_formula = 'dr_visits ~   income + male'

    #demo_socio_formula = 'dr_visits ~ C(marital) + C(race_grp) + income + bmi==1 + bmi==2 + bmi==3 + msa  + male'
    additional_formula = ' + high_bp + high_chol + phy_lim + diabetes + chd'

    dep_2, indep_2 = dmatrices(demo_socio_formula+additional_formula, bmi, return_type = 'dataframe')

    full_poisson_results = sm.GLM(dep_2, indep_2, family=sm.families.Poisson()).fit()
    print('dr_visits',full_poisson_results.summary())
    print('dr_visits',full_poisson_results.aic,'and ', full_poisson_results.bic )


    #dpendent variable: hosp_vis

    demo_socio_formula = 'hosp_vis ~   income  + male'
    #demo_socio_formula = 'hosp_vis ~   C(marital) + C(race_grp) + income +  bmi==1 + bmi==2 + bmi==3 + msa  + male'

    additional_formula = ' + high_bp + high_chol + phy_lim + diabetes + chd'

    dep_2, indep_2 = dmatrices(demo_socio_formula+additional_formula, bmi, return_type = 'dataframe')

    full_poisson_results = sm.GLM(dep_2, indep_2, family=sm.families.Poisson()).fit()
    print('hosp_vis',full_poisson_results.summary())
    print('dr_visits',full_poisson_results.aic,'and ', full_poisson_results.bic )

    # dependent variable: hosp_vis + dr_visits
    demo_socio_formula = 'hlth_util ~   income  + male'
    #demo_socio_formula = 'hlth_util ~   C(marital) + C(race_grp) + income +  bmi==1 + bmi==2 + bmi==3 + msa  + male'

    additional_formula = ' + high_bp + high_chol + phy_lim + diabetes + chd'

    dep_2, indep_2 = dmatrices(demo_socio_formula+additional_formula, bmi, return_type = 'dataframe')

    full_poisson_results = sm.GLM(dep_2, indep_2, family=sm.families.Poisson()).fit()

    print('hosp_vis + dr_visits',full_poisson_results.summary())
    print('dr_visits',full_poisson_results.aic,'and ', full_poisson_results.bic )
    print('---------------------------------------------------------------------------------------------')
```
